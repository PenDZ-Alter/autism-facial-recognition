```python
train = 8497
test = 1967
valid = 1783
total = train + test + valid
```

```python
print(f"Total: {total} data")
print(f"Train: {train} data, {train/total:.2%}")
print(f"Test: {test} data, {test/total:.2%}")
print(f"Valid: {valid} data, {valid/total:.2%}")
```

```python
train_80 = int(total * 0.8)
test_10 = int(total * 0.1)
valid_10 = int(total * 0.1)

print(f"Train: {train_80} data, {train_80/total:.2%}"
      f"\nTest: {test_10} data, {test_10/total:.2%}"
      f"\nValid: {valid_10} data, {valid_10/total:.2%}")
```
