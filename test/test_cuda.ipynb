```python
import tensorflow as tf

gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print("✅ Linker CUDA is Active, Blackwell RTX 5060 Ready!")
        from tensorflow.python.client import device_lib
        print([x.physical_device_desc for x in device_lib.list_local_devices() if x.device_type == 'GPU'])
    except RuntimeError as e:
        print(e)
else:
    print("❌ GPU not detected!")
```

```python
# Check if GPU detected?
import tensorflow as tf
gpus = tf.config.list_physical_devices('GPU')

if gpus:
    for i, gpu in enumerate(gpus):
        # Get Specifications Details
        details = tf.config.experimental.get_device_details(gpu)
        gpu_name = details.get('device_name', 'Unknown GPU')
        compute_cap = details.get('compute_capability', 'Unknown')
        
        print(f"\n--- GPU #{i} ---")
        print(f"Name       : {gpu_name}")
        print(f"ID Path    : {gpu.name}")
        print(f"Compute Cap: {compute_cap}")
        
    # Built and Additional information
    print("\n--- Additional Information ---")
    print(f"TensorFlow Version          : {tf.__version__}")
    print(f"Built with CUDA             : {tf.test.is_built_with_cuda()}")
    print(f"Compute Capabilities Build  : {tf.sysconfig.get_build_info()['cuda_compute_capabilities']}")
else:
    print("❌ GPU Not Detected!")
```

```python
import tensorflow as tf
import ctypes

print(f"--- [ GPU & CUDA STATUS ] ---")
print(f"TensorFlow Version: {tf.__version__}")
print(f"GPU Available: {tf.config.list_physical_devices('GPU')}")
print(f"CUDA Version (TF Build): {tf.sysconfig.get_build_info()['cuda_version']}")

print(f"\n--- [ CHECKING LIBRARIES IN /OPT/CUDA ] ---")

def check_lib(lib_name, name):
    try:
        lib = ctypes.CDLL(lib_name)
        print(f"✅ {name}: is detected!")
        return lib
    except OSError:
        print(f"❌ {name}: not detected. Check LD_LIBRARY_PATH and the Linker!")
        return None

# cuDNN Check
print(f"cuDNN Version (TF Build): {tf.sysconfig.get_build_info()['cudnn_version']}")

# Check any libraries compability
nccl = check_lib("libnccl.so", "NCCL")
nvshmem = check_lib("libnvshmem_host.so", "NVSHMEM")
#nvshmem = check_lib("libnvshmem.so", "NVSHMEM") # Old Version
cublas = check_lib("libcublas.so", "cuBLAS")
curand = check_lib("libcurand.so", "cuRAND")
cusolver = check_lib("libcusolver.so", "cusolver")
cusparse = check_lib("libcusparse.so", "cusparse")
cufft = check_lib("libcufft.so", "cufft")
cudart = check_lib("libcudart.so", "cudart")
```
